## 3. ***Data Wrangling***

````python
#since we saw missing values so we are ready to handle the missing values
# Missing Values/Null Values Count
# Missing values in titles dataset
print("Missing Values in Titles Dataset:\n")

titles_missing = titles.isnull().sum()

print(titles_missing)

print("\n--------------------------------------\n")

# Missing values in credits dataset
print("Missing Values in Credits Dataset:\n")

credits_missing = credits.isnull().sum()

print(credits_missing)
````

````python
#step1-Handle the missing values in Titles Datset


# Fill description (text column)
titles['description'] = titles['description'].fillna(
    "No description available"
)

# Fill age certification (categorical column)
titles['age_certification'] = titles[
    'age_certification'
].fillna(
    "Not Rated"
)

# Fill seasons (numerical but applicable mainly for TV Shows)
titles['seasons'] = titles['seasons'].fillna(0)

# Fill numerical columns using median

titles['imdb_score'] = titles['imdb_score'].fillna(
    titles['imdb_score'].median()
)

titles['imdb_votes'] = titles['imdb_votes'].fillna(
    titles['imdb_votes'].median()
)

titles['tmdb_popularity'] = titles[
    'tmdb_popularity'
].fillna(
    titles['tmdb_popularity'].median()
)

titles['tmdb_score'] = titles['tmdb_score'].fillna(
    titles['tmdb_score'].median()
)


# Step 2 — Handle Missing Values in Credits Dataset

credits['character'] = credits['character'].fillna(
    "Unknown"
)


print("Missing values handled successfully!")
````

````python
#since we saw missing values so we are ready to handle the missing values
# Missing Values/Null Values Count
# Missing values in titles dataset
print("Missing Values in Titles Dataset:\n")

titles_missing = titles.isnull().sum()

print(titles_missing)

print("\n--------------------------------------\n")

# Missing values in credits dataset
print("Missing Values in Credits Dataset:\n")

credits_missing = credits.isnull().sum()

print(credits_missing)
````

#Feature 2: We create Short ,Medium Long Based on Runtime Category

````python
# Create runtime category

def runtime_category(x):

    if x < 60:
        return "Short"

    elif x < 120:
        return "Medium"

    else:
        return "Long"


titles['runtime_category'] = titles['runtime'].apply(runtime_category)

print("Runtime category created successfully.")
````

````python
titles[['runtime','runtime_category']].head()
````

##Feature 3: We categorize IMDb Scores into Low,Average,High

````python

# Feature Engineering — Rating Category

def rating_category(x):

    if x < 5:
        return "Low"

    elif x < 7:
        return "Average"

    else:
        return "High"


titles['rating_category'] = titles[
    'imdb_score'
].apply(rating_category)

print("Rating category created successfully.")
````
